```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure()
```

```python
plt.plot([1,2],[3,4])
```

```python
x= np.arange(1,10)
y= 2*x
```

```python
# Create Figure (empty canvas)
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([.5,.5, 1, 1]) #  xaxis y axis , width and height
axes.plot(x,y)
```

```python
a = np.linspace(0,10,11)
b = a**2
```

```python
figur = plt.figure()
axes = figur.add_axes([.5,.5, 1, 1])
axes.plot(a,b)
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes2  = fig.add_axes([.25,0.25,.25,0.25])
axes1.set_title('large curve')
axes2.set_title('small curve')
axes1.plot(a,b)
axes2.plot(a,b)
```
