```python
def quality(total_bill,tip):
    if tip/total_bill  > 0.25:
        return "Generous"
    else:
        return "Other"
```

```python
#applying the apply function on multiple columns 
df1[['total_bill','tip']].apply(lambda df : quality(df['total_bill'],df['tip']),axis=1)
```

```python
# we can also use np.vectorize function to compute the same task as above, vectorize is much faster than the above lambda execution
np.vectorize(quality)(df1['total_bill'],df1['tip'])
```

```python
import timeit 
  
# code snippet to be executed only once 
setup = '''
import numpy as np
import pandas as pd
df = pd.read_csv('tips.csv')
def quality(total_bill,tip):
    if tip/total_bill  > 0.25:
        return "Generous"
    else:
        return "Other"
'''
  
# code snippet whose execution time is to be measured 
stmt_one = ''' 
df['Tip Quality'] = df[['total_bill','tip']].apply(lambda df: quality(df['total_bill'],df['tip']),axis=1)
'''

stmt_two = '''
df['Tip Quality'] = np.vectorize(quality)(df['total_bill'], df['tip'])
'''
  
```

```python
timeit.timeit(setup = setup, 
                    stmt = stmt_one, 
                    number = 1000) 
```

```python
timeit.timeit(setup = setup, 
                    stmt = stmt_two, 
                    number = 1000) 
```

```python
# MERGING

registrations = pd.DataFrame({'reg_id':[1,2,3,4],'name':['Andrew','Bobo','Claire','David']})
logins = pd.DataFrame({'log_id':[1,2,3,4],'name':['Xavier','Andrew','Yolanda','Bobo']})
```

```python
registrations
```

```python
logins
```

```python
pd.merge(registrations,logins,how='inner',on='name')
```

```python
pd.merge(registrations,logins,how = 'outer',on='name')
```

```python
pd.merge(registrations,logins,how='left',on = 'name')
```

```python
pd.merge(registrations,logins,how='right',on= 'name')
```

```python
# we can also merge the data on the index with the column name in the other table
registrations = registrations.set_index('name')
```

```python
registrations
```

```python
# now i want to merge registrations with the login on index
pd.merge(registrations,logins,left_index=True,right_on='name',how= 'inner')
```

```python
registrations = registrations.reset_index()
```

```python
registrations.columns = ['reg_name','reg_id']
```

```python
# now we can see in both registration and login columns name column got change but have the same data sow how we can do it
pd.merge(registrations,logins,how= 'inner',left_on='reg_name',right_on='name')
```

```python
registrations.columns
```

```python
registrations.columns = ['name','id']
```

```python
logins.columns
```

```python
logins.columns= ['id','name']
```

```python
# if we have same id column in both the table but we merge table on other how python work
pd.merge(registrations,logins,how='inner',on='name')
```

```python
#now in above result python give different name for same id column from different table, we can define suffix 
pd.merge(registrations,logins,how='inner',on='name',suffixes=('_reg','_log'))
```
